Même chose pour read_json et read_excel

```python
base_url ='https://raw.githubusercontent.com/Harlequelrah/Learning_data_science/main/Learning_data_science/data/'
csv_url  = base_url+'clients.csv'
json_url = base_url + 'clients.json'
excel_url = base_url + 'clients.xlsx'
```

```python
# data_csv=pd.read_csv('clients.csv') #  import local
data_csv=pd.read_csv(csv_url)
data_csv.head() # retourne les 5 premiers résultats
```

Pour les fichiers json il s'agit de la ***read_json***

```python
# data_json=pd.read_json('clients.json') # import local
data_json=pd.read_json(json_url)
data_json.head() # retourne les 5 premiers résultats
```

```python
# data_excel = pd.read_excel('clients.xlsx') # import local
data_excel = pd.read_excel(excel_url)
data_excel.head() # retourne les 5 premiers résultats
```
