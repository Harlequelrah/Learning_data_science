# Classifiez les données avec la régression logistique

## Découvrez le principe de régression logistique (ou classification)

Dans un problème de classification, la variable cible est une catégorie et non une valeur continue comme dans la régression.

On a : 

- classification binaire (la variable prend 2 valeurs exclusives)

- classification multiclasse (la variable peut prendre plus de 2 valeurs) : 

  - ordinale 

  - nominale

- classification multilabel (un échantillon peut appartenir à plusieurs catégories à la fois)

## Comprenez le rapport entre classification et régression


Plus précisément, une régression logistique dans un cas de classification binaire de prédiction consiste à :

- modéliser une régression linéaire ;

- projeter la prédiction y variable dans l'intervalle [0,1]: logistique(y) ;

- interpréter logistic(y) comme probabilité que y soit dans une catégorie ou dans l'autre en fonction d'un seuil t=0.5 :

- si logistique(y) < 0.5 => catégorie 0,

- si logistique(y) >= 0.5 => catégorie 1.

Cette fonction de projection de toute valeur réelle dans l'intervalle   [0,1] existe et elle s'appelle la fonction logistique.

D'où l'appellation régression logistique pour parler de l'équivalent de la régression linéaire mais pour la classification.

# Résolvez un problème de classification avec la régression logistique

Prenons un exemple avec le modèle de sklearn LogisticRegression. Considérons le dataset du cancer du sein disponible sur UCI et directement dans sklearn.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression

# Chargement des données
X,y = load_breast_cancer(return_X_y=True)

clf = LogisticRegression(random_state=808)
clf.fit(X,y)
```

```python
# le 8eme echantillon
e1 = [X[8, :]]
#  le 13eme echantillon
e2=[X[13, :]]

display(clf.predict(e1))

display(clf.predict(e2))
```

La prediction pour e1 vaut 0 donc benine

La prediction pour e2 vaut 1 donc maligne

```python
# les probabilités de prédiction pour 0 et 1

# e1
display(clf.predict_proba(e1))
# e2
display(clf.predict_proba(e2))
```

La probabilité de prédiction  de e1 vaut 69% pour 0 et 31% pour 1

La probabilité de prédiction  de e2 vaut 12% pour 0 et 88% pour 1

Traçons l'histogramme des prédictions pour verifier la fiabilité du modèle

```python
import seaborn as sns
import matplotlib.pyplot as plt

y_hat_proba  = clf.predict_proba(X)[:,1]

sns.histplot(y_hat_proba)

plt.title("Histogramme des probabilités")
```

Le modèle est assez fiable au niveau de ses prédictions, la plupart des prédictions ont une probabilité proche de 0 ou de 1.

# Évaluez la performance d'un modèle de classification

```python
# récuperation des prédictions
y_pred = clf.predict(X)
```

```python
# importation des metriques
from sklearn.metrics import accuracy_score

score =accuracy_score(y_pred,y)

print(f"score : {score:.2%}")
```

94.73 % est un très bon score

### Matrice de confusion

Ce modèle est-il biaisé ou objectif ? A-t-il tendance à classer plutôt les 0 en 1 ou les 1 en 0 ?



```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_pred,y)
```

La matrice de confusion se presente comme suite : 

|  |  | Catégorie prédite  |  |
|---|:---:|:---:|:---:|
|  |   | positif | négatif  |
| Catégorie réelle | positif  | TP : vrai positif | FN : faux négatif |
|  | négatif  | FP : faux positif |  TN : vrai négatif|

**note** : scikit-learn a une inversion du positif et du négatif , ce qui donne : 

|  |  | Catégorie prédite  |  |
|---|:---:|:---:|:---:|
|  |   | négatif | positif  |
| Catégorie réelle | négatif  |  TN : vrai négatif|FP : faux positif  |
|  | positif  | FN : faux négatif | TP : vrai positif  |


La diagonale contient le nombre d'échantillons bien classés.

Hors diagonale, le nombre des échantillons mal classés.

où : 

- TP : true positive, vrai positif : échantillons positifs prédits comme positifs.

- FP : false positive, faux positif : échantillons négatifs prédits comme positifs de façon erronée.

- TN : true negative, vrai négatif : échantillons négatifs prédits comme négatifs.

- FN : false negative, faux négatif : échantillons positifs prédits comme négatifs de façon erronée.

## Les métriques

### Le rappel (recall)
Le rappel est adapté pour minimiser les faux négatifs, quand les conséquences de manquer des instances positives sont graves.

Rappel = TP / TP+FN

### La précision (precision)
La précision est adaptée quand on souhaite minimiser les faux positifs.

Precision = TP / TP + FP

| Métrique	| Question posée |	Risque majeur|
|:---:|:---:|:---:|
| Rappel |	"Est-ce que j'ai bien capturé tous les coupables ?" |	Laisser un coupable en liberté (Faux Négatif).|
| Précision | 	"Est-ce que tous ceux que j'ai arrêtés sont vraiment coupables ?" |	Condamner un innocent (Faux Positif).

Les scores obtenus sont dépendants du seuil de classification mentionné auparavant.

Par défaut, on considère un seuil de  t=0.5: si la probabilité prédite   <0.5 alors la prédiction est la catégorie 0, sinon 1. Mais on peut considérer d'autres valeurs pour ce seuil.

```python
from sklearn.metrics import recall_score,precision_score

print(f"recall : {recall_score(y_pred,y):.2%}")

print(f"precision : {precision_score(y_pred,y):.2%}")
```

## ROC

L'idée est de tracer le rappel ou TPR (true positive rate) par rapport au FPR (false positive rate) en fonction des seuils de classification. Le TPR est défini par TP / (TP + FN) et le FPR est défini par FP (FP + TN).

La courbe obtenue est appelée Receiver Operating Characteristic (ou ROC), ou encore fonction d’efficacité du récepteur.

Plus la courbe se rapproche du coin en haut à gauche, meilleur est le modèle. La diagonale correspond à un modèle purement aléatoire.

La métrique qui nous intéresse est en fait la surface entre la courbe et la diagonale. Elle est appelée ROC- AUC pour Area under the ROC ou surface sous la courbe ROC. Elle prend en compte tous les seuils de classement entre 0 et 1 et donc ne dépend pas d'une valeur spécifique de ce seuil.



```python
from sklearn.metrics import roc_curve,roc_auc_score

fpr,tpr,thresholds  = roc_curve(y,y_hat_proba)

plt.plot(fpr,tpr)
plt.plot([0,1],[0,1])

plt.title(f"ROC - Logistic Regression - Breast Cancer Data - ROC-AUC ({roc_auc_score(y_pred,y):.2%})")
```

## Logistic regression sur iris dataset

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# chargement du dataset
X,y = load_iris(return_X_y=True)

# séparation des données
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=8)

# instanciation et entrainement du modèle
clf = LogisticRegression(random_state=8)
clf.fit(X_train,y_train)

# recupération des prédictions
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import  roc_auc_score

roc_auc_score(y_test, clf.predict_proba(X_test),multi_class='ovr')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```
