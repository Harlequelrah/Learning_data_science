# `Évaluez la performance d’un modèle prédictif`


Nous allons : 

- Charger un dataset.

- Entraîner le modèle.

- Évaluer le modèle.

**note** : le dataset comprend l'âge en mois , le sexe en binaire 0 pour garçon et 1 pour fille , taille en centimètres et poids en kg .

Notre modèle essaye de prédire le poids à partir des autres variables. 

On a la forme : 

**poids = a * sexe + b * age + c * taille + bruit**

```python
# importation de pandas
import pandas as pd

# importation du modèle linéaire
from sklearn.linear_model import LinearRegression

# importation de numpy
import numpy as np

# importations des metriques
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python

# chargement du dataframe
df = pd.read_csv('../data/age_vs_poids_vs_taille_vs_sexe.csv')

# affichage d'un extrait du dataframe
df.head()
```

```python
# affichage des dimensions du dataframe
df.shape
```

```python
# les variables predictives
predictives = ['sexe','age','taille']
X = df[predictives]

# la variable cible
target='poids'
y = df[target]

# instanciation du modèle
reg = LinearRegression()

# entrainement
reg.fit(X,y)

# affichage du score
display(reg.score(X,y))

# affichage des coefficients a , b , c
# de la regression lineaire
display(reg.coef_)
```

```python
# garçon de 150 mois et de 153cm
x1_data=[[0, 150, 153]]
x1 = pd.DataFrame(x1_data,columns=predictives)

# son poids
poids = reg.predict(x1)
poids
```

## `Minimiser une fonction de coût`

Le but principal d'un algorithme de Machine Learning est de minimiser l'erreur d'estimation.

En géneralisant l'idée à n'importe quelle fonction permettant de mesurer l'ecart entre les prédictions du modèle et les valeurs réelles de la cible.

## `Évaluez la performance d'un modèle`

La fonction de coût permet de faire tourner l'algorithme d'entraînement du modèle. Mais pour évaluer sa performance une fois entraîné, nous avons besoin de calculer son score de performance.

Le calcul du score d'un modèle dépend de la tâche ML considérée.


```python
# recuperation des predictions du modèle
y_pred = reg.predict(X)

display(mean_squared_error(y,y_pred))

display(mean_absolute_error(y, y_pred))

display(mean_absolute_percentage_error(y, y_pred))
```
