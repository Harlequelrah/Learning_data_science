# Découvrez les notions de modèle et d’algorithme

## Découvrez les différents types de modèles

On choisit le modèle en fonction du contexte et du cahier des charges du projet : 

- tâche à effectuer ;
- volumes des données ;
- nature des données ;
- caractéristiques statistiques des données.

On distingue les classes principales de modèles : 

- Les modèles de Regression;

- Les modèles à base d'arbres;

- Réseaux de neuronnes et le Deep Learning.

## Les modèles de regression
Ils pourront servir à des problèmes de regression ou de classification.
**note** : ne pas confondre modèle de regression et tâche de regression.
