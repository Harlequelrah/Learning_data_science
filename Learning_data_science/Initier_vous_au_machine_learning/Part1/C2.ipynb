# Découvrez les notions de modèle et d’algorithme

## Découvrez les différents types de modèles

On choisit le modèle en fonction du contexte et du cahier des charges du projet : 

- tâche à effectuer ;
- volumes des données ;
- nature des données ;
- caractéristiques statistiques des données.

On distingue les classes principales de modèles : 

- Les modèles de Regression;

- Les modèles à base d'arbres;

- Réseaux de neuronnes et le Deep Learning.


## Les modèles de regression
Ils pourront servir à des problèmes de regression ou de classification.

**note** : ne pas confondre modèle de regression et tâche de regression.

## Les modèles à base d'arbre
Arbre de décision (decision tree), forêt aléatoire (random forest), ainsi que XGBoost et ses nombreuses variantes, ces modèles sont adaptés à des jeux de données tabulaires de taille plus conséquente.Ils sont très robustes face aux données manquantes et aux observations aberrantes 

## Les réseaux de neurones et le Deep Learning

Adaptés à des jeux de données de gros volume, de nature plus complexe, comme des images ou de la vidéo. Bien plus polyvalents dans la nature des données prises en compte (images, vidéo), ils offrent des gains de performances significatifs lorsqu’appliqués à de gros volumes de données.

## Definition d'un modèle

```python
# la classe du modèle
from sklearn.linear_model import LinearRegression

# données artificielles
X = [[0,0],[0,1]]
y = [0,1]

# intanciation du modèle
model = LinearRegression()

# entrainement
model.fit(X,y)
```
