```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
prets = pd.read_csv('https://raw.githubusercontent.com/Harlequelrah/Learning_data_science/refs/heads/main/Learning_data_science/Cr%C3%A9ez_des_data_frames_avec_Pandas/prets.csv')
data = prets.groupby('ville')['remboursement'].sum()
data=data.reset_index()
data
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=data['ville'],height=data['remboursement'])
plt.title("Chiffre d'affaire par agence",fontsize=18,fontname='serif',fontweight='bold')
plt.xlabel("Agences",color='red',fontweight='bold')
plt.ylabel("Chiffre d'affaire (€)")
plt.grid(axis='y')

# plt.ylim(0,100000)
# plt.yticks([0,20000,40000,60000,80000,100000])

ca = data['remboursement'].tolist()
for i in range(6):
  plt.text(i-0.2,ca[i]+2000,round(ca[i],1))
plt.show()
```
