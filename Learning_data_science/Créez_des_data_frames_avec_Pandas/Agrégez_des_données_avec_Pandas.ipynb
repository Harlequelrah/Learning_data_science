En poursuivant avec le contexte de Banque , calculer le chiffre d'affaires mensuel réalisé par chaque agence serait agrégation .

Une agrégation est une opération de l'algèbre linéaire souvent utilisé sur les data frame .

```python
import pandas as pd
import numpy as np
data=pd.read_csv('https://raw.githubusercontent.com/Harlequelrah/Learning_data_science/main/Learning_data_science/Cr%C3%A9ez_des_data_frames_avec_Pandas/prets.csv')
data.head()
```

**Agréger plusieurs lignes**

*Utilisation de groupby de pandas* :

- Choisir les colonnes qui seront les index des résultats

- Choisir la fonction d'agrégation à appliquer

- les colonnes fixées en index se retrouvent en index et pour y accéder on peut avoir besoin de reset_index

```python
data.groupby('ville').sum() # le chiffre d'affaire total de chacune de nos agences
```

```python
data.groupby(['ville','type']).sum() # le chiffre d'affaire total par agence et par type de prêt
```

```python
data.groupby(['ville','type'])['remboursement'].sum()
```

```python
data.groupby('ville').agg({'remboursement':['sum','mean'],'revenu':'max'})
# la moyenne et la somme deremboursement par agence, ainsi que le maximum de revenu
```

Agréger des lignes et des colonnes

La méthode est .pivot_table . Elle prend 4 arguments en paramètres :

index : variable(s) placée(s) en ligne ;

columns : variable(s) placée en colonne(s) ;

values : variable sur laquelle on va appliquer la fonction d’agrégation ;

aggfunc : fonction d’agrégation.

La méthode permettant la transformation inverse s’appelle melt et ses paramètres sont :
     
- id_vars : variable sur laquelle on va agréger les données ;

- value_vars : ensemble des colonnes qu'on souhaite repasser en une seule colonne ;

```python
result = data.pivot_table(
    index='ville',
    columns='type',
    values='remboursement',
    aggfunc='sum'
) # le remboursement par agence selon la ville et le type de prêt
```

```python
result = result.reset_index()
result
```

```python
result.melt(
    id_vars = 'ville',
    value_vars = ['automobile','immobilier']
)
```
