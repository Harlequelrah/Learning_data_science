Ici j'utilise **read_csv** pour lire les données csv de ***clients.csv*** que j'ai au préalable uploader en allant dans fichier et sélectionnant l'option *Importer dans l'espace de stockage de session* .

```python
data_csv=pd.read_csv('clients.csv')
data_csv.head() # retourne les 5 premiers résultats
```

Pour les fichiers json il s'agit de la ***read_json***

```python
data_json=pd.read_json('clients.json')
data_json.head() # retourne les 5 premiers résultats
```

```python
data_excel = pd.read_excel('clients.xlsx')
data_excel.head() # retourne les 5 premiers résultats
```

**Identifiez les caractéristiques de votre data frame**

- la méthode tail d'avoir les n dernière lignes du data frame , n étant 5 par défaut .
